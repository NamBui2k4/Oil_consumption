```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df0 = pd.read_csv('../data/Oil Consumption by Country 1965 to 2023.csv')
```

```python
df0.index = df0.iloc[:,0]
df = df0.T[1:]
df.tail()
```

```python
df.info()
```

```python
stat = df.describe().T

stat.iloc[0:5].plot(kind='bar',rot=True,title='Statistic by sampled countries'),
```

```python
for country in stat.iloc[0:5].index:
    plt.plot(df.index, df[[country]],label=country)
    plt.xticks(ticks=df.index[::5], rotation=45)
    plt.legend()
    plt.title('Trend from 1965 to 2023')
```

```python
check_nan = df.isnull()
ax = plt.figure(figsize=(12,5))
ax = sns.heatmap(check_nan.T)
ax.set_label('NaN observations')
```

```python
df['time'] = pd.to_datetime(df.index)
cols = ['time'] + [col for col in df.columns if col != 'time']
df = df[cols]

# analyze the Iraq consumption
target = 'Iraq'
Iraq_df = df['Iraq'].fillna(0)
time = df['time']

fig, ax = plt.subplots()
ax.plot(time,Iraq_df)
ax.set_xlabel('time')
ax.set_ylabel('oil consumption')
ax.set_title('Trending of oil consumption in Iraq')
```

```python
data_at_2023= df.iloc[-1:,0:-1].T
data_at_2023
```

```python
data_at_2023['2023'] = pd.to_numeric(data_at_2023['2023'], errors='coerce')
top_20 = data_at_2023.sort_values('2023',ascending=False)
top_20
```

```python
nations=top_20.index.tolist()
nations
```

```python
top_nations = [
 'United States',
 'China',
 'India',
 'Russia',
 'Japan',
 'South Korea',
 'Brazil',
 'Canada',
 'Singapore',
 'France',
 'United Kingdom',
 'Spain',
'Italy',
 'Thailand',
 ]
```

```python
plt.figure(figsize=(12,5))
for col in top_nations:
    plt.plot(df.index, df[col], label=col)
    
plt.xlabel('Year')
plt.ylabel('TWh')
plt.title("Trends of tops country",fontsize=20)
plt.xticks(ticks=df.index[::5])
plt.legend()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```

```python
world_21_23 = df['World'].iloc[-4:]
usa_21_23 = df['United States'].iloc[-4:]

data_21_23 = pd.concat([usa_21_23, world_21_23], axis=1).values

fig, ax = plt.subplots(2,2,figsize=(12,5))
ax[0,0].pie(data_21_23[0],labels=['United States','World'],autopct='%1.1f%%')
ax[0, 0].set_title('Year 2020')

ax[0,1].pie(data_21_23[1],labels=['United States','World'],autopct='%1.1f%%')
ax[0, 1].set_title('Year 2021')

ax[1,0].pie(data_21_23[2],labels=['United States','World'],autopct='%1.1f%%')
ax[1,0].set_title('Year 2022')

ax[1,1].pie(data_21_23[3],labels=['United States','World'],autopct='%1.1f%%')
ax[1,1].set_title('Year 2023')

fig.suptitle("USA compared to the World", fontsize=20, x=0.5, y=1.05)
plt.show()
```
